```python
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/chennai-super-kings-squad-1413620/series-squads','CSK'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/delhi-capitals-squad-1413547/series-squads','DC'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/gujarat-titans-squad-1413572/series-squads','GT'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/kolkata-knight-riders-squad-1413553/series-squads','KKR'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/lucknow-super-giants-squad-1413571/series-squads','LSG'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/mumbai-indians-squad-1413556/series-squads','MI'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/punjab-kings-squad-1413565/series-squads','PBKS'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/rajasthan-royals-squad-1413568/series-squads','RR'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/royal-challengers-bangalore-squad-1413548/series-squads','RCB'),
('https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/sunrisers-hyderabad-squad-1413569/series-squads','SRH')
]

# Function to scrape player names from a given URL
def scrape_player_names(url):
    # Send a GET request to the URL
    response = requests.get(url)
    # Initialize BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # The list to store player names
    player_names = []
    
    # List of IDs to look fo
    
    mainContainer = soup.find('div', class_='ds-mb-4').find('div', class_ ='ds-p-0')
    divs = mainContainer.find_all('div', recursive=False)
    
    c1 = divs[0].find('div', class_ = 'ds-grid lg:ds-grid-cols-2', recursive=False).find_all('div', recursive=False)
    for div in c1:
        playername = div.find('div', class_='ds-flex-1').find('a')['title']
        type_ = div.find('div', class_ = 'ds-flex-1').find('div', class_='ds-flex ds-space-x-2').find('span', recursive=False)
        if type_ and type_.text == '†':
            playername = '†'+playername
        player_names.append(playername)
    
    c2 = divs[1].find('div', class_ = 'ds-grid lg:ds-grid-cols-2', recursive=False).find_all('div', recursive=False)
    for div in c2:
        playername = div.find('div', class_='ds-flex-1').find('a')['title']
        type_ = div.find('div', class_ = 'ds-flex-1').find('div', class_='ds-flex ds-space-x-2').find('span', recursive=False)
        if type_ and type_.text == '†':
            playername = '†'+playername
        player_names.append(playername)
        
    c3 = divs[2].find('div', class_ = 'ds-grid lg:ds-grid-cols-2', recursive=False).find_all('div', recursive=False)
    for div in c3:
        playername = div.find('div', class_='ds-flex-1').find('a')['title']
        type_ = div.find('div', class_ = 'ds-flex-1').find('div', class_='ds-flex ds-space-x-2').find('span', recursive=False)
        if type_ and type_.text == '†':
            playername = '†'+playername
        player_names.append(playername)

    return player_names

players_db = {}

# Iterate over each URL and print extracted player names
for tup in urls:
    # get the players in a team
    url = tup[0]
    team = tup[1]
    players_db[team] = {}
    player_names = scrape_player_names(url)
    # initialize them in the dictionary
    for player in player_names:
        if player in players_db[team].keys():
            print("um we have a duplicate player???")
        else:
            players_db[team][player] = 0.0

    print(f"URL: {url}")
    print("Player Names:", player_names)
```

```python
import requests
from bs4 import BeautifulSoup

header = "https://www.espncricinfo.com"
def extract_match_links(url):
    # Send HTTP request to the given URL
    response = requests.get(url)
    
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Initialize a list to store collected links
    collected_links = []
    
    # Find the main container div
    main_container = soup.find('div', class_='ds-mb-4')
    
    # Check if the main container is found
    if not main_container:
        return "Main container not found."
    
    # Find the first child div of main container with no class name
    child_container = main_container.find('div', class_='ds-p-0')
    # get the immediate div children
    divs = child_container.find_all('div', recursive=False)
    print(len(divs))
    for i in range(len(divs)):
        dv = divs[i]
        # Find the div containing the link
        # Find the <a> tag within the link_container
        a_tag = dv.find('a')
        if a_tag:
            # Find the <p> tag within the info_div
            p_tag = a_tag.find('p', class_='ds-text-tight-s ds-font-medium ds-line-clamp-2 ds-text-typo')
            span_tag = p_tag.find('span')
            if span_tag and span_tag.text=='Match yet to begin':
                break
            else:
                lnk = header + a_tag['href']
                collected_links.append(lnk)
        else:
            print("no a tag found??")

    
    return collected_links


links = extract_match_links(url = 'https://www.espncricinfo.com/series/indian-premier-league-2024-1410320/match-schedule-fixtures-and-results')
```

```python
import re
from fuzzywuzzy import process


def returnBattingScore(runs):
    points = float(runs)
    if runs in range(50,75):
        points += float(25)
    elif runs in range(75,100):
        points += float(35)
    elif runs == 100:
         points += float(50)
    return points

def returnBowlingScore(wickets):
    points = 25.0 * float(wickets)
    if wickets == 3:
        points += 25.0
    elif wickets == 4:
        points += 35.0
    elif wickets == 5:
        points += 50.0
    return points


def parse_name(comment):
    if comment[:5] == 'c & b':
        words = comment.split(' ')
        if words[3] == 'sub':
            pname = comment.split('c & b sub ')
            print("pname:")
            pattern = r"\((.*?)\)|\[(.*?)\]"  # Pattern to capture text in brackets or parentheses
            match = re.search(pattern, pname[1])
            if match:
                return tuple(["c",match.group(1)]) if match.group(1) else tuple(["c",match.group(2)])  # Handling both groups
            else:
                return None
        else:
            return tuple(["c",comment[6:]]) #except for c & b return the entire word
    if comment[0] == 'c':
        words = comment.split(' ')
        b_idx = 0
        for word in words:
            if word == 'b':
                break
            b_idx += 1

        # Slice the words list to get all elements before 'b', then join them back into a string
        main_content = ' '.join(words[:b_idx])
        words = main_content.split(' ')
        if words[1] == 'sub':
            pname = comment.split('c sub ')
            pattern = r"\((.*?)\)|\[(.*?)\]"  # Pattern to capture text in brackets or parentheses
            match = re.search(pattern, pname[1])
            if match:
                return tuple(["c",match.group(1)]) if match.group(1) else tuple(["c",match.group(2)])  # Handling both groups
            else:
                return None
        else:
            return tuple(['c',main_content[2:]])
    if comment[:2] == 'st':
        words = comment.split(' ')
        b_idx = 0
        for word in words:
            if word == 'b':
                break
            b_idx += 1
        # Slice the words list to get all elements before 'b', then join them back into a string
        main_content = ' '.join(words[:b_idx])
        words = main_content.split(' ')
        if words[1] == 'sub':
            pname = comment.split('st sub ')
            pattern = r"\((.*?)\)|\[(.*?)\]"  # Pattern to capture text in brackets or parentheses
            match = re.search(pattern, pname[1])
            if match:
                return tuple(["st",match.group(1)]) if match.group(1) else tuple(["st",match.group(2)])  # Handling both groups
            else:
                return None
        else:
            return tuple(["st",main_content[3:]])
        
    return None


def print_stats(arr):
    for t in arr:
        name = getFullName(t[0], players_db[t[1]])
        score = players_db[t[1]][name]
        print(f"{name} | {score}")
        
def getFullName(nickname, team):
    full_names = list(team.keys())
    if '†' in nickname:
        actual_list = [name for name in full_names if '†' in name]
        return process.extractOne(nickname, actual_list)[0] 
    return process.extractOne(nickname, full_names)[0] 



        
        


        
        




    
```

```python

import pandas as pd  # Import pandas


  

def extract_info(url):
    response = requests.get(url)
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # extract the team names so we can update player scores for those teams
    title = soup.find('h1', class_='ds-text-title-xs ds-font-bold ds-mb-2 ds-m-1') 
    teams = title.text.split(',')[0].split('vs')
    team1_name = teams[0].strip()
    team2_name = teams[1].strip()


    print(f"{team1_name} vs {team2_name}")

    # extract the teams' batting and bowling scores
    innings =  soup.find_all('div', class_ = 'ds-rounded-lg ds-mt-2')
    team1_innings = innings[0].find('div', class_='ds-p-0')
    team2_innings = innings[1].find('div', class_='ds-p-0')
    team1_batting_bowling = team1_innings.find_all('table', recursive=False)
    team2_batting_bowling = team2_innings.find_all('table', recursive=False) 
    
    team1_batting = team1_batting_bowling[0]
    team1_bowling = team1_batting_bowling[1]
    team2_batting = team2_batting_bowling[0]
    team2_bowling = team2_batting_bowling[1]

    team1_batting_rows = team1_batting.find('tbody').find_all('tr', recursive=False)
    team1_batting_rows = team1_batting_rows[0:(len(team1_batting_rows)-4)] #ignore the last 4 rows which contain the Fall of wickets, 
                   
    team1_bowling_rows = team1_bowling.find('tbody').find_all('tr', recursive=False)


    team2_batting_rows = team2_batting.find('tbody').find_all('tr', recursive=False)
    team2_batting_rows = team2_batting_rows[0:(len(team2_batting_rows)-4)] #ignore the last 4 rows which contain the Fall of wickets, 
                                          # players who did not bat, total runs, and extras
    team2_bowling_rows = team2_bowling.find('tbody').find_all('tr', recursive=False)
    
    for tr in team1_batting_rows:
        if tr['class'] == []: #then we have a player record
            tds = tr.find_all('td', recursive=False)
            b_player_name = tds[0].find('a')['title']
            outCommentary = tds[1].find_all('span')
            full_player_name = None
            type_ = ""
            if len(outCommentary) != 0:
               commentary = outCommentary[-1].text.strip()
               info = parse_name(comment=commentary)
               if info is not None:
                    type_ = info[0]
                    nickname = info[1]
                    full_player_name = getFullName(nickname,players_db[team2_name])
                    if type_ == 'c':
                       players_db[team2_name][full_player_name] += 3.0
                    if type_ == 'st':
                       players_db[team2_name][full_player_name] += 5.0
                       
            runs = int(tds[2].find('strong').text)
            
            score = returnBattingScore(runs)
            pname = getFullName(b_player_name,players_db[team1_name])
            players_db[team1_name][pname] += score
           
    for tr in team1_bowling_rows:
       if tr['class'] == []:
          tds = tr.find_all('td', recursive=False)
          nameContainer = tds[0]
          bowler_name = nameContainer.find('span').text.strip()
          wicketsContainer = tds[4]
          wickets = int(wicketsContainer.find('strong').text.strip())
          score = returnBowlingScore(wickets)
          pname = getFullName(bowler_name, players_db[team2_name])
          players_db[team2_name][pname] += score
          
    for tr in team2_batting_rows:
        if tr['class'] == []: #then we have a player record
            tds = tr.find_all('td', recursive=False)
            b_player_name = tds[0].find('a')['title']
            outCommentary = tds[1].find_all('span')
            full_player_name = None
            type_ = ""
            if len(outCommentary) != 0:
               commentary = outCommentary[-1].text.strip()
               info = parse_name(comment=commentary)
               if info is not None:
                type_ = info[0]
                nickname = info[1]
                nickname = nickname[1:] if nickname[0] == '†' else nickname
                full_player_name = getFullName(nickname,players_db[team1_name])
                if type_ == 'c':
                       players_db[team1_name][full_player_name] += 3.0
                if type_ == 'st':
                       players_db[team1_name][full_player_name] += 5.0
            
            runs = int(tds[2].find('strong').text)
            score = returnBattingScore(runs)
            pname = getFullName(b_player_name,players_db[team2_name])
            players_db[team2_name][pname] += score
    
    for tr in team2_bowling_rows:
       if tr['class'] == []:
          tds = tr.find_all('td', recursive=False)
          nameContainer = tds[0]
          bowler_name = nameContainer.find('span').text.strip()
          wicketsContainer = tds[4]
          wickets = int(wicketsContainer.find('strong').text.strip())
          score = returnBowlingScore(wickets)
          pname = getFullName(bowler_name, players_db[team1_name])
          players_db[team1_name][pname] += score
          


    



    
```
